```python
import pandas as pd
import numpy as np
# Load the dataset
file_path = 'daily_data.csv'
data = pd.read_csv(file_path)

# Print the column names and first few rows to inspect the data
# print(data.columns)
print(data.head())
```

```python
day_ids = data['day_id']
day_ids
```

```python
data['city_id'] = data['city_id'].str.replace('C', '').astype(int)
```

```python
# Fill missing values with column means for numeric data
modified_columns = ['city_id','temperature_celsius','condition_text', 'wind_kph', 'wind_degree', 'pressure_mb',
    'precip_mm', 'humidity', 'cloud', 'feels_like_celsius',
    'visibility_km', 'uv_index', 'gust_kph', 'air_quality_us-epa-index'
]
# Display the first few rows to verify the changes
numerical_data = data[modified_columns]
numerical_data.isnull().sum()
```

```python
numerical_data.loc[numerical_data['condition_text'] == '', 'condition_text'] = 'null'
```

```python
numerical_data['condition_text'] = numerical_data['condition_text'].replace(r'^\s*$', np.nan, regex=True)
numerical_data
```

```python
# Convert 'sunrise' and 'sunset' to datetime format
data['sunrise'] = pd.to_datetime(data['sunrise'], format='%I:%M %p').dt.time
data['sunset'] = pd.to_datetime(data['sunset'], format='%I:%M %p').dt.time

# Calculate daylight duration in minutes
data['sunrise'] = pd.to_datetime(data['sunrise'].astype(str))
data['sunset'] = pd.to_datetime(data['sunset'].astype(str))
numerical_data['daylight_duration'] = (data['sunset'] - data['sunrise']).dt.seconds / 60
```

```python
numerical_data
```

```python
weather_mapping = {
    'Clear and Sunny': 1,
    'Partly Cloudy': 2,
    'Light Precipitation': 3,
    'Cloudy and Overcast': 4,
    'Mist or Fog': 5,
    'Rain Showers': 6,
    'Light Rain with Thunder': 7,
    'Thunderstorms': 8,
    'Moderate to Heavy Rain': 9
}
```

```python
numerical_data['condition_text'] = numerical_data['condition_text'].str.strip()
numerical_data
```

```python
# Replace categories with numerical values in 'condition_text' column
numerical_data['condition_text'] = numerical_data['condition_text'].replace(weather_mapping)
numerical_data
```

```python
numerical_data.head(10)
```

```python
training_data = numerical_data.drop(columns=["condition_text"])
target_data = numerical_data["condition_text"]
target_data[target_data.isna()] = -1
target_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(training_data)
training_data = scaler.transform(training_data)
training_data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.semi_supervised import SelfTrainingClassifier
base_classifier = RandomForestClassifier(random_state=42)
self_training_model = SelfTrainingClassifier(base_classifier)
self_training_model.fit(training_data, target_data)
```

```python
predictions = self_training_model.predict(training_data)
predictions
```

```python
# Trainig and testing data
from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(training_data, predictions, test_size=0.3, random_state=100)
X_train, X_test, y_train, y_test = train_test_split(training_data, predictions, test_size=0.2, random_state=100)
```

```python
# Train a Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=100, n_estimators=100, max_depth=200, min_samples_split=2, min_samples_leaf=1)
clf.fit(X_train, y_train)
```

```python
# Predict the labels for the test set
y_pred = clf.predict(X_test)

# Calculate the accuracy of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
# actual vs predicted
actual_vs_predicted = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
actual_vs_predicted
```

```python
# Performance metrics
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# Predict the weather condition
predictions = clf.predict(training_data)
pred_df = pd.DataFrame({'Predicted': predictions})
pred_df
```

```python
# Convert the condition_text column back to the original categories
reverse_mapping = {1: 'Clear and Sunny', 2: 'Partly Cloudy', 3: 'Light Precipitation', 4: 'Cloudy and Overcast',
                   5: 'Mist or Fog', 6: 'Rain Showers', 7: 'Light Rain with Thunder', 8: 'Thunderstorms',
                   9: 'Moderate to Heavy Rain'}
pred_df['Predicted'] = pred_df['Predicted'].replace(reverse_mapping)
pred_df
```

```python
# Read submission file and update the condition_text column
submission_file = 'submission.csv'
submission_data = pd.read_csv(submission_file)
submission_data['condition_text'] = pred_df['Predicted']
submission_data.head(20)
```

```python
# Save the updated submission file
submission_data.to_csv('submission_p220_4.csv', index=False)
```
