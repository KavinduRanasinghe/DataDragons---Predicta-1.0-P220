```python
import pandas as pd
import numpy as np
# Load the dataset
file_path = 'daily_data.csv'
data = pd.read_csv(file_path)

# Print the column names and first few rows to inspect the data
# print(data.columns)
print(data.head())
```

```python
day_ids = data['day_id']
day_ids
```

```python
data['city_id'] = data['city_id'].str.replace('C', '').astype(int)
```

```python
# Fill missing values with column means for numeric data
modified_columns = ['city_id','temperature_celsius','condition_text', 'wind_kph', 'wind_degree', 'pressure_mb',
    'precip_mm', 'humidity', 'cloud', 'feels_like_celsius',
    'visibility_km', 'uv_index', 'gust_kph', 'air_quality_us-epa-index'
]
# Display the first few rows to verify the changes
numerical_data = data[modified_columns]
numerical_data.isnull().sum()
```

```python
numerical_data.loc[numerical_data['condition_text'] == '', 'condition_text'] = 'null'
```

```python
numerical_data['condition_text'] = numerical_data['condition_text'].replace(r'^\s*$', np.nan, regex=True)
numerical_data
```

```python
# Convert 'sunrise' and 'sunset' to datetime format
data['sunrise'] = pd.to_datetime(data['sunrise'], format='%I:%M %p').dt.time
data['sunset'] = pd.to_datetime(data['sunset'], format='%I:%M %p').dt.time

# Calculate daylight duration in minutes
data['sunrise'] = pd.to_datetime(data['sunrise'].astype(str))
data['sunset'] = pd.to_datetime(data['sunset'].astype(str))
numerical_data['daylight_duration'] = (data['sunset'] - data['sunrise']).dt.seconds / 60
```

```python
numerical_data
```

```python
weather_mapping = {
    'Clear and Sunny': 1,
    'Partly Cloudy': 2,
    'Light Precipitation': 3,
    'Cloudy and Overcast': 4,
    'Mist or Fog': 5,
    'Rain Showers': 6,
    'Light Rain with Thunder': 7,
    'Thunderstorms': 8,
    'Moderate to Heavy Rain': 9
}

# Strip leading and trailing spaces from 'condition_text' column


# Display the updated DataFrame


```

```python
numerical_data['condition_text'] = numerical_data['condition_text'].str.strip()
numerical_data
```

```python
# Replace categories with numerical values in 'condition_text' column
numerical_data['condition_text'] = numerical_data['condition_text'].replace(weather_mapping)
numerical_data
```

```python
numerical_data.head(10)
```

```python
original_numerical_data = numerical_data
original_numerical_data
```

```python
# Remove rows with missing values
numerical_data = numerical_data.dropna()
numerical_data
```

```python
# Normalize the data without 'condition_text' column using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
target_variable = numerical_data['condition_text'].astype(int)
numerical_data = numerical_data.drop('condition_text', axis=1)
numerical_data_scaled = scaler.fit_transform(numerical_data)
numerical_data_scaled = pd.DataFrame(numerical_data_scaled, columns=numerical_data.columns)
numerical_data_scaled
```

```python
# Trainig and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(numerical_data_scaled, target_variable, test_size=0.2, random_state=10)
```

```python
# Neural network model for classification for weather condition. condition_text is the target variable.
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
model = MLPClassifier(hidden_layer_sizes=(20, 50, 20), max_iter=1000, random_state=0, activation='relu', solver='adam', alpha=0.0001, batch_size='auto', learning_rate='constant', learning_rate_init=0.001, power_t=0.5, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True, early_stopping=False, validation_fraction=0.1, beta_1=0.9, beta_2=0.999, epsilon=1e-08, n_iter_no_change=10, max_fun=15000)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
accuracy
```

```python
# Check accuaracy for different models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier()
}
for model_name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    print(f'{model_name} accuracy: {accuracy}')
```

```python
# actual vs predicted
actual_vs_predicted = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
actual_vs_predicted.head(20)
```
